```python
import pandas as pd
```

```python
content_similarity = pd.read_pickle('item_item_content_similarity_matrix.pickle')
rating_similarity = pd.read_pickle('item_item_rating_similarity_matrix.pickle')
```

```python
content_similarity
```

```python
rating_similarity
```

```python
content_similarity = content_similarity.loc[rating_similarity.columns.to_list()].sort_index(axis=1)
content_similarity = content_similarity.reindex(rating_similarity.columns.to_list())
content_similarity = content_similarity[rating_similarity.columns.to_list()] 
```

```python
content_similarity
```

```python
similarity_matrix = content_similarity + rating_similarity

similarity_matrix
```

```python
def sort_column_with_names(column):
    sorted_values = column.sort_values(ascending=False)
    sorted_tuples = [(value, column_name) for column_name, value in sorted_values.items()]
    return sorted_tuples

# Apply the function to each column
sorted_similarity = similarity_matrix.apply(sort_column_with_names, axis=1, result_type='expand')
```

```python
sorted_similarity
```

```python
result = sorted_similarity.iloc[:, :5]
```

```python
result
```

```python
def extract_second_value(column):
    return [value[1] for value in column]

result = result.apply(extract_second_value)
```

```python
result
```
