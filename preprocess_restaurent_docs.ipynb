```python
import pickle
import pandas as pd
```

```python
with open('houston_rest_df.pickle', 'rb') as f:
    business_df = pickle.load(f)
```

```python
with open('houston_restaurent_reviews.pickle', 'rb') as f:
    reviews_df = pickle.load(f)
```

```python
reviews_df.head(3)
```

```python
business_df.head(3)
```

```python
merged_business_review_df = pd.merge(business_df, reviews_df, on='gmap_id', how='left')
```

```python
merged_business_review_df.head(4)
```

```python
merged_business_review_df = merged_business_review_df.dropna(subset=['text'])
```

```python
merged_review_df = merged_business_review_df.groupby('gmap_id')['text'].apply(lambda x: ' '.join(x)).reset_index()
```

```python
merged_review_df.head(5)
```

```python
doc_df = pd.merge(business_df, merged_review_df, on='gmap_id', how='left')
```

```python
doc_df['doc_information'] = doc_df['description'].fillna(' ') + ' ' + doc_df['category'].apply(lambda x: ', '.join(x) if isinstance(x, list) else str(x)) + ' ' + doc_df['MISC'].apply(lambda x: ', '.join(val for val in x.values()) if isinstance(x, dict) else '')
```

```python
def concat_dict_values(d):
    if d is None:
        return ''
    return ', '.join([f"{' '.join(value)}" for key, value in d.items()])


doc_df['doc_information'] = doc_df['description'].fillna(' ') + ' ' + doc_df['category'].apply(lambda x: ' '.join(x) if isinstance(x, list) else str(x)) + ' ' + doc_df['MISC'].apply(concat_dict_values) + doc_df['text']
```

```python
doc_df.head(3)
```

```python
import pickle
with open('restaurent_docs.pickle', 'wb') as f:
    pickle.dump(doc_df, f)
```
